## Ensure consistency

````python

def calculate_label_percentages(y_labels):
    """
    Calculates and prints the percentage of 'normal' (1) and 'bad' (0) instances in the given labels.

    Parameters:
    y_labels (array-like): The dataset labels.

    Returns:
    None
    """
    # Count the occurrences of each class
    unique, counts = np.unique(y_labels, return_counts=True)
    class_counts = dict(zip(unique, counts))
    
    # Calculate percentages
    total_counts = len(y_labels)
    percentage_normal = (class_counts[1] / total_counts) * 100 if 1 in class_counts else 0
    percentage_bad = (class_counts[0] / total_counts) * 100 if 0 in class_counts else 0
    
    # Print the results
    print("Total number of samples:", total_counts)
    print(f"Percentage of 'normal' (1): {percentage_normal:.2f}%")
    print(f"Percentage of 'bad' (0): {percentage_bad:.2f}%")
````

## Training Baseline Model with Floating-Point Precision

We are training a baseline model using the CatBoostClassifier with the following settings:
- **Model Type**: `float`
- **Tree Size**: 100 iterations
- **Learning Rate**: 0.5
- **Depth**: 3 layers

The trained model is then saved in JSON format at the specified location

````python

model_type='float'
Tree_size=100
lr=0.5
depth=3
model_cat = CatBoostClassifier(
    iterations=Tree_size,
    learning_rate=lr,
    depth=depth,verbose=99
)
model_cat.fit(
    X_train_part, y_train_part,
    #eval_set=(X_validation, y_validation),
    verbose=10
)

model_path='models_json/'
# Create the extraction directory if it doesn't exist
os.makedirs(model_path, exist_ok=True)

model_float=model_path+f"{model_type}_T{Tree_size}_lr{lr}_d{depth}.json"
model_cat.save_model(model_float, format="json")
````

## Model Performance Evaluation

We evaluate the classification performance of the trained CatBoost model primarily focusing on the AUC (Area Under the Curve) as it provides a measure of performance across all classification thresholds. Optimal threshold could be found from validation set, but here, the validation set is small.

````python


y_pred_classes = model_cat.predict(X_test, prediction_type='RawFormulaVal')
#print(y_pred_classes)
y_pred_prob = expit(y_pred_classes)

y_pred = (y_pred_prob >= 0.8).astype(int)

accuracy=accuracy_score(y_test,y_pred)

fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc=auc(fpr, tpr)

# Optimal threshold is the one where the sum of sensitivity (TPR) and specificity (1 - FPR) is maximized
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]
print('Optimal threshold = ',optimal_threshold)
print('Accuracy = ',accuracy)
# Use the optimal threshold for classification
y_pred_optimal = (y_pred_prob >= optimal_threshold).astype(int)

# Calculate accuracy with the optimal threshold
accuracy_optimal = accuracy_score(y_test, y_pred_optimal)
print('Optimal Accuracy',accuracy_optimal)
print('AUC',roc_auc)
````

## Quantization-Aware Training Configuration for GBDT Model

Our GBDT model is configured to utilize quantization-aware training to optimize its deployment on specific hardware setups:

- **Quantization Flags**:
  - **Thresholds**: Configured to be unsigned and symmetric.
  - **Leaves**: Configured to be unsigned and symmetric.
**Note** Unsigned is not supported yet.

- **Training Parameters**:
  - **Precision** (`N`): N-bit precision for quantization.
  - **Number of Iterations Before Quantization** (`NIQ`): Set as Hyperparameter to help with Loss decreasing.
  - **Number of Trees** (`treesize`).
  - **Depth of Trees** (`depth`).
  - **Learning Rate** (`lr`).
  - **If BinaryGBDT** (`Binary`): Set to True if training Binary GBDT.


````python
Threshold_is_unsigned= False
Threshold_is_symetric= False
Leaf_is_unsigned= False
Leaf_is_symetric= True
N_iterations_to_quantise= 4


Flag = [Threshold_is_unsigned, Threshold_is_symetric, Leaf_is_unsigned, Leaf_is_symetric]
N=8
treesize=100
depth=3
lr=0.5


# Start time
start_time = time.time()

model_name, model1, model2, learn_losses_sign_8_4, test_losses,S_threshold_per_tree8,Z_threshold_per_tree8,S_leaf_per_tree8, Z_leaf_per_tree8 = qa_train_model(
    X_train_part, y_train_part,X_validation,
    y_validation,X_test,y_test,
    Flag=Flag,
    N=N,
    NIQ=N_iterations_to_quantise,
    learning_rate=lr,depth=depth, tree_size=treesize,
    Binary=False)

# End time
end_time = time.time()

# Calculate the elapsed time
elapsed_time = end_time - start_time
print(f"Training time: {elapsed_time:.2f} seconds")
````

## Training a Binary GBDT Model

When training a Binary GBDT model, ensure the training data is appropriately prepared and the model parameters are accurately configured.

### Data Preparation:
- **Binary Scaling**: The training data must be scaled to the range of (-1, 1) instead of the typical (0, 1).

### Model Configuration:
- **Quantization Flags**: configuring them will not affect training resultes since they are not related to the concept of Binary training.
- **Training Parameters**:
  - **Precision** (`N`): Set to 1 to reflect the binary nature of the model.
  - **Number of Trees** (`treesize`)
  - **Depth of Trees** (`depth`)
  - **Learning Rate** (`lr`)
  - **Iterations Before Quantization** (`NIQ`) 
  - **Binary Mode**: Ensured to be `True`.
### Important note for Trainig with numpy.sing() for leaf values:
- The sign function returns -1 if x < 0, 0 if x==0, 1 if x > 0. nan is returned for nan inputs.
- The training will proceed better np.sing() compared to using np.where(x<=0, -1, 1)
- After finishing the training, we used np.where(x<=0, -1, 1) to have only (-1,1) instead of (-1,0,1), and achived the same accuracy.

````python
Threshold_is_unsigned= False
Threshold_is_symetric= False
Leaf_is_unsigned= False
Leaf_is_symetric= False
N_iterations_to_quantise= 2


Flag = [Threshold_is_unsigned, Threshold_is_symetric, Leaf_is_unsigned, Leaf_is_symetric]
N=1
treesize=100
depth=3
lr=0.5
# Start time
start_time = time.time()

model_name, model1, model2, learn_losses_sign_b_2, test_losses,S_threshold_per_treeb,Z_threshold_per_treeb,S_leaf_per_treeb, Z_leaf_per_treeb = qa_train_model(
    Xb_train_part, yb_train_part,Xb_validation,
    yb_validation,Xb_test,yb_test,
    Flag=Flag,
    N=N,
    NIQ=N_iterations_to_quantise,
    learning_rate=lr,depth=depth, tree_size=treesize,
    Binary=True)

# End time
end_time = time.time()

# Calculate the elapsed time
elapsed_time = end_time - start_time
print(f"Training time: {elapsed_time:.2f} seconds")
````

## Comparing Model Performance: QAT vs. PTQ

Assessing the performance differences between models using Quantization-Aware Training (QAT) and Post-Training Quantization (PTQ)

````python


# Configuration settings
model_type = 'float'
Tree_size = 100
lr = 0.5
depth = 3
Ns = range(2, 17,2)  # Range of integer bits

# Evaluate model with PTQ GBDT
model_float = model_path+f"{model_type}_T{Tree_size}_lr{lr}_d{depth}.json"
_ , _,auc_non_quantized = evaluate_model_performance_without_quantization(model_float, X_test, y_test)


Threshold_is_unsigned= False
Threshold_is_symetric= False
Leaf_is_unsigned= False
Leaf_is_symetric= True

Flag = [Threshold_is_unsigned, Threshold_is_symetric, Leaf_is_unsigned, Leaf_is_symetric]

auc_list_quantized = []
for N in Ns:
    model_integer_post = model_path+f"post_integer_T{Tree_size}_lr{lr}_d{depth}_N{N}.json"
    # Update quantization parameters
    S_threshold, Z_threshold, S_leaf, Z_leaf = update_to_PTQ(model_float, model_integer_post, Flag=Flag, N=N)

    # Evaluate performance with quantization
    _, _ , auc_quantized = evaluate_model_performance(model_integer_post, X_test, y_test, S_threshold, Z_threshold, S_leaf, Z_leaf, N, Flag)

    auc_list_quantized.append(auc_quantized)
# Reset the color cycle again
#plt.gca()
# Plot Accuracy Rates vs N
plt.figure(figsize=(3.5, 3.5), dpi=300)
# plt.rc('font', family='Times New Roman', size=14)
plt.plot(list(Ns), auc_list_quantized, 'o-', label='PTQ-int (2 to 16)',markeredgecolor='black')
plt.xlabel('N (Integer Bits)', fontsize=14)
plt.ylabel('AUC', fontsize=14)
plt.title('AUC vs Integer Bits (N)', fontsize=14)
plt.xticks(list(Ns), fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)


#=======================================================================================================================

Threshold_is_unsigned= False
Threshold_is_symetric= False
Leaf_is_unsigned= False
Leaf_is_symetric= True

Flag = [Threshold_is_unsigned, Threshold_is_symetric, Leaf_is_unsigned, Leaf_is_symetric] 
N=8
treesize=100
depth=3
lr=0.5
model_type='QA_integer'
model_integer_QA=model_path+f"{model_type}_T{treesize}_lr{lr}_d{depth}_N{N}.json"

data = np.load(model_path+'my_tree_S_and_Z.npz')
# Extract arrays from the loaded data
S_threshold_per_tree8 = data['S_threshold_per_tree8']
Z_threshold_per_tree8 = data['Z_threshold_per_tree8']
S_leaf_per_tree8 = data['S_leaf_per_tree8']
Z_leaf_per_tree8 = data['Z_leaf_per_tree8']


# Add data point for model_integer_QA
accuracy_QA, avg_auc_QA, roc_auc_QA = evaluate_model_performance(model_integer_QA, X_test, y_test, S_threshold_per_tree8[-1:][0], Z_threshold_per_tree8[-1:][0], S_leaf_per_tree8[-1:][0], Z_leaf_per_tree8[-1:][0], N, Flag)

plt.plot(8, roc_auc_QA, '^', markersize=10, alpha=0.75, label=f'QAT-int{N}', markeredgewidth=1, markeredgecolor='black')

#=======================================================================================================================


Threshold_is_unsigned= False
Threshold_is_symetric= False
Leaf_is_unsigned= False
Leaf_is_symetric= False

Flag = [Threshold_is_unsigned, Threshold_is_symetric, Leaf_is_unsigned, Leaf_is_symetric]
N=1
treesize=100
depth=3
model_type='Binary'
lr=0.5
model_integer_Binary=model_path+f"{model_type}_T{Tree_size}_lr{lr}_d{depth}_N{N}.json"
data = np.load(model_path+'my_tree_S_and_Z.npz')
# Extract arrays from the loaded data
S_threshold_per_treeb = data['S_threshold_per_treeb']
Z_threshold_per_treeb = data['Z_threshold_per_treeb']
S_leaf_per_treeb = data['S_leaf_per_treeb']
Z_leaf_per_treeb = data['Z_leaf_per_treeb']


# Add data points for model_Binary
accuracy_Binary, avg_auc_Binary, roc_auc_Binary = evaluate_model_performance_without_quantization(model_integer_Binary, Xb_test, yb_test)

plt.plot(1, roc_auc_Binary, '*', markersize=10, alpha=0.75, label='BinaryGBDT', markeredgewidth=1, markeredgecolor='black')

#=======================================================================================================================


plt.xticks(np.arange(0, 17, 2))  # This will set ticks at 0, 2, 4, ..., 32

# Adjust plot properties
plt.xlim(0, 17)  # Set the x-axis limits to show from 0 to 17
plt.tick_params(axis='both', which='major', labelsize=10)

ax = plt.gca()

ax.axhline(y=auc_non_quantized, color='gray', linestyle='--', label=f'Float')
# Updated legend parameters
ax.legend(loc='lower right', fancybox=False, shadow=False, ncol=1, prop={'size': 8}, framealpha=0.5)

plt.tight_layout()

# Save the figure
plt.savefig('AUC_comp_figure_cyper.png', format='png', bbox_inches='tight')
plt.savefig('AUC_comp_figure_cyper.pdf', format='pdf', bbox_inches='tight')

plt.show()

````
